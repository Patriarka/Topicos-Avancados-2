# Introdução

O eletroencefalograma (EEG) é um registro da atividade elétrica do cérebro a partir do couro cabeludo. As formas de onda registradas refletem a atividade elétrica cortical.

Tomamos conhecimento de algumas propriedades:

**(1) Intensidade do sinal**: a atividade do EEG é bastante pequena, medida em microvolts (mV).

**(2) Frequência do sinal**: as principais frequências das ondas do EEG humano são:

- **Delta ($\delta$)**: tem uma frequência de 3 Hz ou menos. Tende a ter a amplitude mais alta e as ondas mais lentas. É normal como ritmo dominante em lactentes até um ano e nos estágios 3 e 4 do sono. Pode ocorrer focalmente com lesões subcorticais e em distribuição geral com lesões difusas, encefalopatia metabólica, hidrocefalia ou lesões profundas da linha média.
- **Teta ($\theta$)**: tem frequência de 3,5 a 7,5 Hz e é classificado como atividade uma "lenta". É normal em crianças de até 13 anos e durante o sono, mas anormal em adultos acordados. Pode ser vista como uma manifestação de lesões subcorticais focais; também pode ser observada em distribuição generalizada em distúrbios difusos, como encefalopatia metabólica ou alguns casos de hidrocefalia.
- **Alfa ($\alpha$)**: tem uma frequência entre 7,5 e 13 Hz. Geralmente é melhor visto nas regiões posteriores da cabeça de cada lado, sendo maior em amplitude no lado dominante. Aparece ao fechar os olhos e relaxar, e desaparece ao abrir os olhos ou um alerta como pensar ou calcular. É o ritmo principal observado em adultos relaxados normais. Está presente durante a maior parte da vida, especialmente após o décimo terceiro ano.
- **Beta ($\beta$)**: a atividade beta é uma atividade "rápida". Tem uma frequência de 14 Hz e superior. Geralmente é visto em ambos os lados em distribuição simétrica e é mais evidente frontalmente. É acentuado por drogas sedativo-hipnóticas, especialmente os benzodiazepínicos e os barbitúricos. Pode estar ausente ou reduzido em áreas de dano cortical. Geralmente é considerado um ritmo normal. É o ritmo dominante em pacientes alertas, ansiosos ou com os olhos abertos.

## Variáveis utilizadas na classificação de atividades com EEG

### Frequência

A frequência refere-se à atividade repetitiva rítmica (em Hz). A frequência da atividade EEG pode ter propriedades diferentes, incluindo:

- **Rítmico**. Atividade de EEG que consiste em ondas de frequência aproximadamente constante.
- **Arrítmico**. Atividade de EEG em que não há ritmos estáveis.
- **Disrítmico**. Ritmos e/ou padrões de atividade de EEG que aparecem em alguns grupos de pacientes ou raramente vistos em indivíduos saudáveis.

### Tensão

A tensão refere-se à tensão média ou tensão de pico da atividade EEG. Os valores dependem, em parte, da técnica de gravação. Os termos descritivos associados à tensão do EEG incluem:

1. **Atenuação (sinônimos: supressão, depressão)**. Redução da amplitude da atividade do EEG resultante da diminuição da tensão. Quando a atividade é atenuada pela estimulação, diz-se que foi "bloqueada" ou que apresenta "bloqueio".
2. **Hipersincronia**. Visto como um aumento na voltagem e regularidade da atividade rítmica, ou dentro da faixa alfa, beta ou teta. O termo implica num aumento do número de elementos neurais que contribuem para o ritmo.
3. **Paroxística**. Atividade que emerge do "*background*" com um início rápido, atingindo (geralmente) uma tensão bastante alta e terminando com um retorno abrupto à atividade de tensão mais baixa. Embora o termo não implique diretamente em anormalidade, muitas atividades anormais são paroxísticas.

### Morfologia

A morfologia refere-se à forma de onda. A forma de uma onda ou padrão de EEG é determinada pelas frequências que se combinam para formar a forma de onda e por suas relações de fase e tensão. Os padrões de onda podem ser descritos como sendo:

- **Monomórfico**. Atividade EEG distinta parecendo ser composta por uma atividade dominante
- **Polimórfico**. atividade EEG distinta composta por múltiplas frequências que se combinam para formar uma forma de onda complexa.
- **Sinusoidal**. Ondas semelhantes a ondas senoidais. A atividade monomórfica geralmente é senoidal.
- **Transitório**. Uma onda ou padrão isolado que é distintamente diferente da atividade de fundo.

    a) ***Spike***: um transiente com pico pontiagudo e duração de 20 - <70 ms.
    b) **Onda aguda**: transiente com pico pontiagudo e duração de 70-200 mseg.

### Sincronia

Sincronia refere-se ao aparecimento simultâneo de padrões rítmicos ou morfologicamente distintos em diferentes regiões da cabeça, seja do mesmo lado (unilateral) ou de ambos os lados (bilateral).

### Periodicidade

A periodicidade refere-se à distribuição de padrões ou elementos no tempo (por exemplo, o aparecimento de uma determinada atividade de EEG em intervalos mais ou menos regulares). A atividade pode ser generalizada, focal ou lateralizada.

## Artefatos

O maior desafio com o monitoramento de EEG é o reconhecimento e eliminação de artefatos. Existem artefatos relacionados ao paciente (por exemplo, movimentação da cabeça, suor, ECG, movimentos oculares) e artefatos técnicos (artefato 50/60 Hz, movimentos do cabo, relacionados à pasta de eletrodos), que devem ser tratados de maneira diferente. Existem algumas ferramentas para encontrar os artefatos. Por exemplo, fEMG e medições de impedância podem ser usadas para indicar sinal contaminado. Observando diferentes parâmetros em um monitor, outras interferências podem ser encontradas.
 
Os eletrodos usados na gravação de EEG não discriminam os sinais elétricos que recebem. A atividade registrada que não é de origem cerebral é denominada artefato e pode ser dividida em fisiológica (gerada pelo sujeito a partir de outras fontes que não o cérebro) e artefatos extrafisiológicos provenientes de fora do corpo (equipamentos incluindo os eletrodos e o ambiente).

### Atividade eletromiografica (EMG)

Atividade EMG são artefatos comuns: os potenciais miogênicos gerados nos músculos frontalis (levantar as sobrancelhas) e os músculos temporais (apertar os músculos da mandíbula) são de duração mais curta do que aqueles gerados no cérebro. Esses artefatos podem ser identificados com base na duração, morfologia e taxa de disparo (frequência). Padrões particulares de artefatos de EMG podem ocorrer em alguns distúrbios do movimento: tremor essencial e doença de Parkinson podem produzir formas de onda sinusoidais rítmicas de 4 a 6 Hz.

### Movimento dos olhos

O globo ocular atua como um dipolo com um pólo positivo orientado anteriormente (córnea) e um pólo negativo orientado posteriormente (retina). Quando o globo gira em torno de seu eixo, ele gera um campo de corrente alternada de grande amplitude detectável por qualquer um dos eletrodos posicionados próximo ao olho. Uma piscada faz com que o polo positivo (a córnea) se aproxime dos eletrodos frontopolares `Fp1`, `Fp2`, produzindo deflexões descendentes simétricas.

### Artefatos da pele

Uma outra dificuldade surge devido às propriedades de certas camadas da pele. Existe um potencial significativo entre o estrato córneo e o estrato granuloso e qualquer deformação local da pele alterará esse potencial. A única maneira confiável de eliminar a fonte do artefato é criar um caminho de baixa resistência através das camadas da pele por meio da limpeza da pele (cotonete com álcool). Além disso, o cloreto de sódio (eletrólito) do suor reagindo com os metais dos eletrodos pode produzir um desvio lento da linha de base.

### Artefato em 50/60 Hz

O problema surge quando a impedância de um dos eletrodos ativos se torna significativamente grande entre os eletrodos e o terra do amplificador. Nessa situação, o solo torna-se um eletrodo que, dependendo de sua localização, produz o artefato de 60 Hz. A interferência da radiação de alta frequência de outros dispositivos eletrônicos pode sobrecarregar os amplificadores de EEG.

### Correções:

- **Amplicação diferencial**: É a chave para o equipamento eletrofisiológico. Amplia a diferença entre duas entradas. Um sinal indesejado comum às duas entradas será subtraído.

- **Filtragem temporal**: As configurações de filtragem padrão para EEG de rotina são:
    - Filtro de baixa frequência: 1 Hz
    - Filtro de alta frequência: 50-70 Hz
